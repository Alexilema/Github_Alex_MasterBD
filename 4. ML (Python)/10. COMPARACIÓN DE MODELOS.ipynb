Explorando sesgo-varianza a través de validación cruzada repetida y boxplot

```python
def boxplots_algorithms(results, names):
    plt.figure(figsize=(12,6))
    plt.boxplot(results)
    plt.title('Comparación final de modelos', fontsize=14)
    plt.xlabel('Modelos', fontsize=14)
    plt.ylabel('CV Score (f1)', fontsize=14)
    plt.xticks(range(1,len(names)+1), names, fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()
```

```python
import pickle
import matplotlib.pyplot as plt

# Nombres de los modelos
names = ["1-RLOG", "2-DTree", "3-BTree",
         "4-RF", "5-GB", "6-XG",
         "7-Red", "8-SVC", "9-RBF",
         "10-BLog", "11-Stack"]

# Rutas a tus puntuaciones guardadas
scores_paths = ["cv_scores_model1.pkl", "cv_scores_model2.pkl", "cv_scores_model3.pkl",
                "cv_scores_model4.pkl", "cv_scores_model5.pkl", "cv_scores_model6.pkl",
                "cv_scores_model7.pkl", "cv_scores_model8.pkl", "cv_scores_model9.pkl",
                "cv_scores_model10.pkl", "cv_scores_model11.pkl"]

# Lista para guardar los resultados
cv_scores = []

for path in scores_paths:
    # Carga las puntuaciones
    with open(path, 'rb') as f:
        scores = pickle.load(f)
    
    # Guarda los resultados
    cv_scores.append(scores)

# Crear el boxplot
boxplots_algorithms(cv_scores, names)


```
